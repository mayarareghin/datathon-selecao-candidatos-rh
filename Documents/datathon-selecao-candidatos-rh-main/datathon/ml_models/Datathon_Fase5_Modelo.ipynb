### V1. Modelo que compara vagas e candidatos e cria um ranking dos candidatos mais compativeis com as vagas


```python
# Instalação do SBERT e tqdm
!pip install -q sentence-transformers tqdm

# Importações
import time

import pandas as pd
from sentence_transformers import SentenceTransformer, util
from sklearn.metrics import ndcg_score
from tqdm import tqdm

# Carregamento dos dados
df_vagas = pd.read_json('vagas.json').T.reset_index(names="id_vaga")
df_vagas = pd.json_normalize(df_vagas.to_dict(orient="records"))

df_applicants = pd.read_json('applicants.json').T.reset_index(names="id_candidato")
df_applicants = pd.json_normalize(df_applicants.to_dict(orient="records"))

# Pré-processa os currículos (aumentado para 1500 caracteres)
df_applicants['cv_pt'] = df_applicants['cv_pt'].fillna('').str.slice(0, 1500)

# Carrega modelo SBERT mais robusto
modelo_sbert = SentenceTransformer('all-mpnet-base-v2')

# Pré-computa embeddings dos candidatos
textos_candidatos = df_applicants['cv_pt'].tolist()
embeddings_candidatos = modelo_sbert.encode(textos_candidatos)


# Função de ranqueamento por bloco com filtro opcional por área
def ranquear_em_blocos(df_vagas, embeddings_candidatos, bloco_size=500, area=None):
    todos_resultados = []
    inicio = time.time()

    # Filtra candidatos por área de atuação, se especificado
    if area:
        candidatos_filtrados = df_applicants[df_applicants['informacoes_profissionais.area_atuacao'] == area]
        textos = candidatos_filtrados['cv_pt'].tolist()
        ids = candidatos_filtrados['id_candidato']
        embeddings = modelo_sbert.encode(textos)
    else:
        ids = df_applicants['id_candidato']
        embeddings = embeddings_candidatos

    for i in range(0, len(df_vagas), bloco_size):
        df_bloco = df_vagas.iloc[i:i + bloco_size]
        resultados = []

        for _, vaga in tqdm(df_bloco.iterrows(), total=len(df_bloco), desc=f"Bloco {i}–{i + bloco_size}"):
            descricao = vaga.get('perfil_vaga.demais_observacoes', '')
            if pd.isna(descricao) or descricao.strip() == '':
                continue

            emb_vaga = modelo_sbert.encode(descricao)
            similaridades = util.cos_sim(emb_vaga, embeddings)[0].tolist()

            ranking = pd.DataFrame({
                'id_vaga': vaga['id_vaga'],
                'id_candidato': ids,
                'similaridade': similaridades
            }).sort_values(by='similaridade', ascending=False)

            resultados.append(ranking)

        bloco_resultado = pd.concat(resultados, ignore_index=True)
        todos_resultados.append(bloco_resultado)

        # Salvamento automático por bloco
        bloco_resultado.to_csv(f'ranking_bloco_{i}_{i + bloco_size}.csv', index=False)

    fim = time.time()
    print(f"\nTempo total de execução: {fim - inicio:.2f} segundos (~{(fim - inicio) / 60:.2f} minutos)")

    return pd.concat(todos_resultados, ignore_index=True)


# Executa o ranqueamento completo (sem filtro de área)
ranking_final = ranquear_em_blocos(df_vagas, embeddings_candidatos)

# Exibe os top 5 candidatos por vaga
top5_por_vaga = ranking_final.groupby('id_vaga').head(5)
print(top5_por_vaga)
```

Avaliação com Métricas de Ranking

```python

def avaliar_modelo(ranking_final, df_prospeccoes, k=5):
    metricas = []

    for id_vaga in ranking_final['id_vaga'].unique():
        # Top K candidatos ranqueados para a vaga
        top_k = ranking_final[ranking_final['id_vaga'] == id_vaga].head(k)

        # Busca a prospecção correspondente pelo id_prospeccao
        linha = df_prospeccoes[df_prospeccoes['id_prospeccao'] == id_vaga]

        candidatos_reais = []
        if not linha.empty and 'prospects' in linha.columns:
            candidatos_reais = [
                str(p['codigo']) for p in linha.iloc[0]['prospects'] if 'codigo' in p
            ]

        # Gera vetor de relevância binária
        relevancias = [1 if str(cid) in candidatos_reais else 0 for cid in top_k['id_candidato']]

        # Calcula métricas
        ndcg = ndcg_score([relevancias], [top_k['similaridade'].tolist()])
        precision = sum(relevancias) / k

        metricas.append({
            'id_vaga': id_vaga,
            'precision@5': precision,
            'ndcg@5': ndcg
        })

    return pd.DataFrame(metricas)
```

```python
# Avaliação
df_metricas = avaliar_modelo(ranking_final, df_prospeccoes)

# Estatísticas gerais
print("\nDesempenho do Modelo:")
print(df_metricas.describe())
```

### **V2. Modelo** supervisionado RandomForest para ranquear candidatos por vaga


```python
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

# Carrega modelo SBERT
modelo_sbert = SentenceTransformer('all-mpnet-base-v2')


# Gera dataset supervisionado
def gerar_dataset_supervisionado(df_vagas, df_applicants, df_prospeccoes, max_negativos=5):
    dados = []

    for _, vaga in tqdm(df_vagas.iterrows(), total=len(df_vagas), desc="Gerando pares supervisionados"):
        id_vaga = vaga['id_vaga']
        descricao = vaga.get('perfil_vaga.demais_observacoes', '')
        if pd.isna(descricao) or descricao.strip() == '':
            continue

        emb_vaga = modelo_sbert.encode(descricao)
        linha = df_prospeccoes[df_prospeccoes['id_prospeccao'] == id_vaga]

        candidatos_reais = []
        if not linha.empty and 'prospects' in linha.columns:
            candidatos_reais = [str(p['codigo']) for p in linha.iloc[0]['prospects'] if 'codigo' in p]

        # Positivos
        for cid in candidatos_reais:
            candidato = df_applicants[df_applicants['id_candidato'].astype(str) == cid]
            if not candidato.empty:
                texto_cv = candidato.iloc[0]['cv_pt'][:1500]
                emb_cv = modelo_sbert.encode(texto_cv)
                dados.append({
                    'embedding': np.concatenate([emb_vaga, emb_cv]),
                    'label': 1
                })

        # Negativos
        candidatos_negativos = df_applicants[~df_applicants['id_candidato'].astype(str).isin(candidatos_reais)].sample(n=max_negativos)
        for _, candidato in candidatos_negativos.iterrows():
            texto_cv = candidato['cv_pt'][:1500]
            emb_cv = modelo_sbert.encode(texto_cv)
            dados.append({
                'embedding': np.concatenate([emb_vaga, emb_cv]),
                'label': 0
            })

    return pd.DataFrame(dados)


# Treinamento do modelo supervisionado
df_supervisionado = gerar_dataset_supervisionado(df_vagas, df_applicants, df_prospeccoes)
X = np.stack(df_supervisionado['embedding'].values)
y = df_supervisionado['label'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rf.fit(X_train, y_train)

# Avaliação
y_pred = modelo_rf.predict(X_test)
print("\nRelatório de classificação:")
print(classification_report(y_test, y_pred))


# Aplicação do modelo supervisionado para ranqueamento
def ranquear_com_modelo_supervisionado(df_vagas, df_applicants, modelo_sbert, modelo_supervisionado):
    resultados = []

    for _, vaga in tqdm(df_vagas.iterrows(), total=len(df_vagas), desc="Ranqueando candidatos por vaga"):
        id_vaga = vaga['id_vaga']
        descricao = vaga.get('perfil_vaga.demais_observacoes', '')
        if pd.isna(descricao) or descricao.strip() == '':
            continue

        emb_vaga = modelo_sbert.encode(descricao)

        for _, candidato in df_applicants.iterrows():
            texto_cv = candidato['cv_pt'][:1500]
            emb_cv = modelo_sbert.encode(texto_cv)
            vetor = np.concatenate([emb_vaga, emb_cv])
            prob = modelo_supervisionado.predict_proba([vetor])[0][1]

            resultados.append({
                'id_vaga': id_vaga,
                'id_candidato': candidato['id_candidato'],
                'probabilidade': prob
            })

    df_ranking = pd.DataFrame(resultados)
    df_ranking = df_ranking.sort_values(by=['id_vaga', 'probabilidade'], ascending=False)
    return df_ranking


# Gera ranking supervisionado
ranking_supervisionado = ranquear_com_modelo_supervisionado(df_vagas, df_applicants, modelo_sbert, modelo_rf)

# Exibe os top 5 candidatos por vaga
top5_supervisionado = ranking_supervisionado.groupby('id_vaga').head(5)
print(top5_supervisionado)
```

```python
# Instalação de pacotes
!pip install -q sentence-transformers xgboost tqdm joblib

# Importações
import math

import joblib
import numpy as np
import pandas as pd
from sentence_transformers import CrossEncoder, SentenceTransformer
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier

# Carregamento dos dados
df_vagas = pd.read_json('vagas.json').T.reset_index(names="id_vaga")
df_vagas = pd.json_normalize(df_vagas.to_dict(orient="records"))

df_applicants = pd.read_json('applicants.json').T.reset_index(names="id_candidato")
df_applicants = pd.json_normalize(df_applicants.to_dict(orient="records"))
df_applicants['cv_pt'] = df_applicants['cv_pt'].fillna('').str.slice(0, 1500)

df_prospeccoes = pd.read_json('prospects.json').T.reset_index(names="id_prospeccao")
df_prospeccoes = pd.json_normalize(df_prospeccoes.to_dict(orient="records"))

# Modelos
modelo_sbert = SentenceTransformer('all-mpnet-base-v2')
modelo_cross = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')


# Codifica features estruturadas
def codificar_features(candidato):
    area = str(candidato.get('informacoes_profissionais.area_atuacao', '')).lower()
    nivel = str(candidato.get('informacoes_profissionais.nivel_profissional', '')).lower()
    ingles = str(candidato.get('formacao_e_idiomas.nivel_ingles', '')).lower()
    escolaridade = str(candidato.get('formacao_e_idiomas.nivel_escolaridade', '')).lower()
    experiencia = candidato.get('informacoes_profissionais.tempo_experiencia', 0)
    return [
        hash(area) % 1000 / 1000,
        hash(nivel) % 1000 / 1000,
        hash(ingles) % 1000 / 1000,
        hash(escolaridade) % 1000 / 1000,
        min(experiencia, 50) / 50
    ]


# Pré-processa candidatos
df_applicants['embedding_cv'] = df_applicants['cv_pt'].apply(lambda x: modelo_sbert.encode(x))
df_applicants['features'] = df_applicants.apply(lambda c: codificar_features(c), axis=1)
df_applicants['area'] = df_applicants['informacoes_profissionais.area_atuacao'].fillna('').str.lower()


# Gera dataset supervisionado
def gerar_dataset_supervisionado(df_vagas, df_applicants, df_prospeccoes, max_negativos=20):
    dados = []

    for _, vaga in tqdm(df_vagas.iterrows(), total=len(df_vagas), desc="Gerando pares supervisionados"):
        id_vaga = vaga['id_vaga']
        descricao = vaga.get('perfil_vaga.demais_observacoes', '')
        area_vaga = str(vaga.get('perfil_vaga.area_atuacao', '')).lower()
        if pd.isna(descricao) or descricao.strip() == '':
            continue

        emb_vaga = modelo_sbert.encode(descricao)
        linha = df_prospeccoes[df_prospeccoes['id_prospeccao'] == id_vaga]
        candidatos_reais = [str(p['codigo']) for p in linha.iloc[0]['prospects']] if not linha.empty else []

        candidatos_area = df_applicants[df_applicants['area'] == area_vaga]

        for cid in candidatos_reais:
            candidato = candidatos_area[candidatos_area['id_candidato'].astype(str) == cid]
            if not candidato.empty:
                emb_cv = candidato.iloc[0]['embedding_cv']
                features = candidato.iloc[0]['features']
                dados.append({'embedding': np.concatenate([emb_vaga, emb_cv, features]), 'label': 1})

        candidatos_negativos = candidatos_area[~candidatos_area['id_candidato'].astype(str).isin(candidatos_reais)].sample(n=max_negativos)
        for _, candidato in candidatos_negativos.iterrows():
            emb_cv = candidato['embedding_cv']
            features = candidato['features']
            dados.append({'embedding': np.concatenate([emb_vaga, emb_cv, features]), 'label': 0})

    return pd.DataFrame(dados)


# Treinamento do modelo supervisionado
df_supervisionado = gerar_dataset_supervisionado(df_vagas, df_applicants, df_prospeccoes)
X = np.stack(df_supervisionado['embedding'].values)
y = df_supervisionado['label'].values

peso_positivo = (y == 0).sum() / (y == 1).sum()
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
modelo_xgb = XGBClassifier(n_estimators=100, eval_metric='logloss', scale_pos_weight=peso_positivo)
modelo_xgb.fit(X_train, y_train)

# Avaliação
y_pred = modelo_xgb.predict(X_test)
print("\nRelatório de classificação:")
print(classification_report(y_test, y_pred))

# Salva modelo supervisionado
joblib.dump(modelo_xgb, 'modelo_supervisionado_xgb.joblib')


# Ranqueamento supervisionado completo
def ranquear_supervisionado(df_vagas, df_applicants, modelo_sbert, modelo_supervisionado, threshold=0.55):
    resultados = []
    embeddings_cv = np.stack(df_applicants['embedding_cv'].values)
    features_cv = np.stack(df_applicants['features'].values)
    ids_cv = df_applicants['id_candidato'].values
    areas_cv = df_applicants['area'].values

    for _, vaga in tqdm(df_vagas.iterrows(), total=len(df_vagas), desc="Ranqueando supervisionado"):
        id_vaga = vaga['id_vaga']
        descricao = vaga.get('perfil_vaga.demais_observacoes', '')
        area_vaga = str(vaga.get('perfil_vaga.area_atuacao', '')).lower()
        if pd.isna(descricao) or descricao.strip() == '':
            continue

        emb_vaga = modelo_sbert.encode(descricao)
        mask_area = areas_cv == area_vaga
        emb_cv_area = embeddings_cv[mask_area]
        features_area = features_cv[mask_area]
        ids_area = ids_cv[mask_area]

        emb_vaga_replicado = np.repeat([emb_vaga], len(emb_cv_area), axis=0)
        vetores = np.concatenate([emb_vaga_replicado, emb_cv_area, features_area], axis=1)
        probs = modelo_supervisionado.predict_proba(vetores)[:, 1]

        for cid, prob in zip(ids_area, probs):
            if prob >= threshold:
                resultados.append({'id_vaga': id_vaga, 'id_candidato': cid, 'probabilidade': prob})

    return pd.DataFrame(resultados)


ranking_supervisionado = ranquear_supervisionado(df_vagas, df_applicants, modelo_sbert, modelo_xgb)


# Reranqueamento semântico
def reranquear_top_candidatos(ranking_supervisionado, df_vagas, df_applicants, modelo_cross, top_n=50, top_k=5):
    resultados_finais = []

    for id_vaga, grupo in tqdm(ranking_supervisionado.groupby('id_vaga'), desc="Reranking semântico"):
        vaga = df_vagas[df_vagas['id_vaga'] == id_vaga].iloc[0]
        descricao = vaga.get('perfil_vaga.demais_observacoes', '')
        if pd.isna(descricao) or descricao.strip() == '':
            continue

        top_candidatos = grupo.sort_values(by='probabilidade', ascending=False).head(top_n)
        pares = []
        ids = []

        for _, linha in top_candidatos.iterrows():
            cid = linha['id_candidato']
            texto = df_applicants[df_applicants['id_candidato'] == cid]['cv_pt'].values[0]
            pares.append((descricao, texto))
            ids.append(cid)

        scores = modelo_cross.predict(pares)
        rerank = sorted(zip(ids, scores), key=lambda x: x[1], reverse=True)[:top_k]

        for cid, score in rerank:
            resultados_finais.append({
                'id_vaga': id_vaga,
                'id_candidato': cid,
                'score_rerank': score
            })

    ranking_final = pd.DataFrame(resultados_finais)
    joblib.dump(ranking_final, 'ranking_final_rerank.joblib')
    return ranking_final


ranking_final = reranquear_top_candidatos(ranking_supervisionado, df_vagas, df_applicants, modelo_cross)


# Métricas de ranking com tratamento de exceções
def dcg(relevancias):
    return sum((2**rel - 1) / math.log2(idx + 2) for idx, rel in enumerate(relevancias))


def ndcg(relevancias):
    ideal = sorted(relevancias, reverse=True)
    return dcg(relevancias) / dcg(ideal) if dcg(ideal) > 0 else 0


def calcular_metricas_ranking(ranking, df_prospeccoes, k=5):
    precisions, recalls, ndcgs = [], [], []

    for id_vaga, grupo in ranking.groupby('id_vaga'):
        linha = df_prospeccoes[df_prospeccoes['id_prospeccao'] == id_vaga]
        if linha.empty or 'prospects' not in linha.columns:
            continue

        encaminhados_reais = set(str(p['codigo']) for p in linha.iloc[0]['prospects'])
        if len(encaminhados_reais) == 0:
            continue  # ignora vagas sem encaminhamentos reais

        relevantes = [1 if str(cid) in encaminhados_reais else 0 for cid in grupo.sort_values(by='score_rerank', ascending=False)['id_candidato'].head(k)]
        precisions.append(sum(relevantes) / k)
        recalls.append(sum(relevantes) / len(encaminhados_reais))
        ndcgs.append(ndcg(relevantes))

    return {
        'Precision@K': round(np.mean(precisions), 3),
        'Recall@K': round(np.mean(recalls), 3),
        'NDCG@K': round(np.mean(ndcgs), 3),
        'Vagas avaliadas': len(precisions)
    }


# Avaliação final
metricas = calcular_metricas_ranking(ranking_final, df_prospeccoes, k=5)
print("\nMétricas de Ranking (K=5):")
for metrica, valor in metricas.items():
    print(f"{metrica}: {valor}")
```

Validação Ranking

```python
def validar_ranking(ranking_final, df_prospeccoes, df_applicants, k=5):
    print("Iniciando validação do ranking...\n")

    # 1. Verifica se há candidatos no ranking
    if ranking_final.empty:
        print("O ranking final está vazio.")
        return

    # 2. Verifica se os IDs de candidatos estão como string
    ranking_final['id_candidato'] = ranking_final['id_candidato'].astype(str)

    # 3. Verifica se há interseção entre candidatos recomendados e encaminhados reais
    total_vagas = 0
    vagas_com_match = 0
    total_matches = 0

    for id_vaga, grupo in ranking_final.groupby('id_vaga'):
        linha = df_prospeccoes[df_prospeccoes['id_prospeccao'] == id_vaga]
        if linha.empty or 'prospects' not in linha.columns:
            continue

        encaminhados_reais = set(str(p['codigo']) for p in linha.iloc[0]['prospects'] if 'codigo' in p)
        if len(encaminhados_reais) == 0:
            continue

        candidatos_preditos = set(grupo.sort_values(by='score_rerank', ascending=False)['id_candidato'].head(k))
        intersecao = encaminhados_reais & candidatos_preditos

        total_vagas += 1
        if intersecao:
            vagas_com_match += 1
            total_matches += len(intersecao)

    print(f"Vagas avaliadas: {total_vagas}")
    print(f"Vagas com pelo menos 1 match: {vagas_com_match}")
    print(f"Total de candidatos corretos no Top-{k}: {total_matches}")

    if vagas_com_match == 0:
        print("\nNenhum candidato relevante foi recuperado no Top-K.")
        print("Sugestões:")
        print("- Verifique se os IDs de candidatos estão alinhados com os 'codigo' dos prospects.")
        print("- Reduza o threshold do ranqueamento supervisionado (ex: de 0.55 para 0.45).")
        print("- Verifique se os currículos estão suficientemente informativos.")
        print("- Teste manualmente uma vaga com candidatos reais e veja se aparecem no ranking.")
    else:
        print("\nRanking está recuperando candidatos relevantes.")
        print("Você pode prosseguir com o cálculo das métricas.")


# Executa o validador
validar_ranking(ranking_final, df_prospeccoes, df_applicants, k=5)
```

# **V8. Script - Versão final- Ranqueamento Inteligente (Supervisionado + Semântico)**

```python
import joblib
import numpy as np
import pandas as pd
from sentence_transformers import CrossEncoder, SentenceTransformer

# Modelos
modelo_sbert = SentenceTransformer('all-mpnet-base-v2')
modelo_cross = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')
modelo_rf = joblib.load('modelo_supervisionado_rf.joblib')

# Dados
df_vagas = pd.read_json('vagas.json').T.reset_index(names="id_vaga")
df_vagas = pd.json_normalize(df_vagas.to_dict(orient="records"))

df_applicants = pd.read_json('applicants.json').T.reset_index(names="id_candidato")
df_applicants = pd.json_normalize(df_applicants.to_dict(orient="records"))
df_applicants['cv_pt'] = df_applicants['cv_pt'].fillna('').str.slice(0, 1500)

df_prospeccoes = pd.read_json('prospects.json').T.reset_index(names="id_prospeccao")
df_prospeccoes = pd.json_normalize(df_prospeccoes.to_dict(orient="records"))


# Features estruturadas
def codificar_features(candidato):
    area = str(candidato.get('informacoes_profissionais.area_atuacao', '')).lower()
    nivel = str(candidato.get('informacoes_profissionais.nivel_profissional', '')).lower()
    ingles = str(candidato.get('formacao_e_idiomas.nivel_ingles', '')).lower()
    escolaridade = str(candidato.get('formacao_e_idiomas.nivel_escolaridade', '')).lower()
    experiencia = candidato.get('informacoes_profissionais.tempo_experiencia', 0)
    return [
        hash(area) % 1000 / 1000,
        hash(nivel) % 1000 / 1000,
        hash(ingles) % 1000 / 1000,
        hash(escolaridade) % 1000 / 1000,
        min(experiencia, 50) / 50
    ]


# Pré-processamento
df_applicants['embedding_cv'] = df_applicants['cv_pt'].apply(lambda x: modelo_sbert.encode(x))
df_applicants['features'] = df_applicants.apply(lambda c: codificar_features(c), axis=1)
df_applicants['area'] = df_applicants['informacoes_profissionais.area_atuacao'].fillna('').str.lower()
df_applicants['id_candidato'] = df_applicants['id_candidato'].astype(str)


# Ranqueamento supervisionado em blocos
def ranquear_supervisionado_em_blocos(df_vagas, df_applicants, modelo_sbert, modelo_supervisionado, threshold=0.55, bloco=500):
    resultados = []
    embeddings_cv = np.stack(df_applicants['embedding_cv'].values)
    features_cv = np.stack(df_applicants['features'].values)
    ids_cv = df_applicants['id_candidato'].values
    areas_cv = df_applicants['area'].values

    total_vagas = len(df_vagas)
    for inicio in tqdm(range(0, total_vagas, bloco), desc="Ranqueando supervisionado"):
        fim = min(inicio + bloco, total_vagas)
        df_bloco = df_vagas.iloc[inicio:fim]

        for _, vaga in df_bloco.iterrows():
            id_vaga = vaga['id_vaga']
            descricao = vaga.get('perfil_vaga.demais_observacoes', '')
            area_vaga = str(vaga.get('perfil_vaga.area_atuacao', '')).lower()
            if pd.isna(descricao) or descricao.strip() == '':
                continue

            emb_vaga = modelo_sbert.encode(descricao)
            mask_area = areas_cv == area_vaga
            emb_cv_area = embeddings_cv[mask_area]
            features_area = features_cv[mask_area]
            ids_area = ids_cv[mask_area]

            if len(emb_cv_area) == 0:
                continue

            emb_vaga_replicado = np.repeat([emb_vaga], len(emb_cv_area), axis=0)
            vetores = np.concatenate([emb_vaga_replicado, emb_cv_area, features_area], axis=1)
            probs = modelo_supervisionado.predict_proba(vetores)[:, 1]

            for cid, prob in zip(ids_area, probs):
                if prob >= threshold:
                    resultados.append({'id_vaga': id_vaga, 'id_candidato': cid, 'probabilidade': prob})

    return pd.DataFrame(resultados)


ranking_supervisionado = ranquear_supervisionado_em_blocos(df_vagas, df_applicants, modelo_sbert, modelo_rf)


# Reranqueamento semântico em blocos
def reranquear_em_blocos(ranking_supervisionado, df_vagas, df_applicants, modelo_cross, top_n=50, top_k=5, bloco=500):
    resultados_finais = []
    vagas = ranking_supervisionado['id_vaga'].unique()
    total_vagas = len(vagas)

    for inicio in tqdm(range(0, total_vagas, bloco), desc="Reranking semântico"):
        fim = min(inicio + bloco, total_vagas)
        vagas_bloco = vagas[inicio:fim]

        for id_vaga in vagas_bloco:
            grupo = ranking_supervisionado[ranking_supervisionado['id_vaga'] == id_vaga]
            vaga = df_vagas[df_vagas['id_vaga'] == id_vaga].iloc[0]
            descricao = vaga.get('perfil_vaga.demais_observacoes', '')
            if pd.isna(descricao) or descricao.strip() == '':
                continue

            top_candidatos = grupo.sort_values(by='probabilidade', ascending=False).head(top_n)
            pares, ids = [], []

            for _, linha in top_candidatos.iterrows():
                cid = linha['id_candidato']
                try:
                    texto = df_applicants[df_applicants['id_candidato'] == cid]['cv_pt'].values[0]
                    if pd.isna(texto) or texto.strip() == '':
                        continue
                    pares.append((descricao, texto))
                    ids.append(cid)
                except Exception as e:
                    print(f"Erro ao processar candidato {cid}: {e}")
                    continue

            try:
                scores = modelo_cross.predict(pares)
                rerank = sorted(zip(ids, scores), key=lambda x: x[1], reverse=True)[:top_k]
                for cid, score in rerank:
                    resultados_finais.append({'id_vaga': id_vaga, 'id_candidato': cid, 'score_rerank': score})
            except Exception as e:
                print(f"Erro no reranking da vaga {id_vaga}: {e}")
                continue

    ranking_final = pd.DataFrame(resultados_finais)
    joblib.dump(ranking_final, 'ranking_final_rerank.joblib')
    return ranking_final


ranking_final = reranquear_em_blocos(ranking_supervisionado, df_vagas, df_applicants, modelo_cross)


# Métricas de ranking
def dcg(relevancias):
    return sum((2**rel - 1) / math.log2(idx + 2) for idx, rel in enumerate(relevancias))


def ndcg(relevancias):
    ideal = sorted(relevancias, reverse=True)
    return dcg(relevancias) / dcg(ideal) if dcg(ideal) > 0 else 0


def calcular_metricas_ranking(ranking, df_prospeccoes, k=5):
    precisions, recalls, ndcgs = [], [], []

    for id_vaga, grupo in ranking.groupby('id_vaga'):
        linha = df_prospeccoes[df_prospeccoes['id_prospeccao'] == id_vaga]
        if linha.empty or 'prospects' not in linha.columns:
            continue

        encaminhados_reais = set(str(p['codigo']) for p in linha.iloc[0]['prospects'] if 'codigo' in p)
        if len(encaminhados_reais) == 0:
            continue

        relevantes = [
            1 if str(cid) in encaminhados_reais else 0
            for cid in grupo.sort_values(by='score_rerank', ascending=False)['id_candidato'].head(k)
        ]
        precisions.append(sum(relevantes) / k)
        recalls.append(sum(relevantes) / len(encaminhados_reais))
        ndcgs.append(ndcg(relevantes))

    return {
        'Precision@K': round(np.mean(precisions), 3),
        'Recall@K': round(np.mean(recalls), 3),
        'NDCG@K': round(np.mean(ndcgs), 3),
        'Vagas avaliadas': len(precisions)
    }


metricas = calcular_metricas_ranking(ranking_final, df_prospeccoes, k=5)

# Exibe os resultados
print("\nMétricas de Ranking (K=5):")
for metrica, valor in metricas.items():
    print(f"{metrica}: {valor}")
```

```python
def validar_ranking(ranking_final, df_prospeccoes, df_applicants, k=5):
    print("Iniciando validação do ranking...\n")

    # 1. Verifica se há candidatos no ranking
    if ranking_final.empty:
        print("O ranking final está vazio.")
        return

    # 2. Verifica se os IDs de candidatos estão como string
    ranking_final['id_candidato'] = ranking_final['id_candidato'].astype(str)

    # 3. Verifica se há interseção entre candidatos recomendados e encaminhados reais
    total_vagas = 0
    vagas_com_match = 0
    total_matches = 0

    for id_vaga, grupo in ranking_final.groupby('id_vaga'):
        linha = df_prospeccoes[df_prospeccoes['id_prospeccao'] == id_vaga]
        if linha.empty or 'prospects' not in linha.columns:
            continue

        encaminhados_reais = set(str(p['codigo']) for p in linha.iloc[0]['prospects'] if 'codigo' in p)
        if len(encaminhados_reais) == 0:
            continue

        candidatos_preditos = set(grupo.sort_values(by='score_rerank', ascending=False)['id_candidato'].head(k))
        intersecao = encaminhados_reais & candidatos_preditos

        total_vagas += 1
        if intersecao:
            vagas_com_match += 1
            total_matches += len(intersecao)

    print(f"Vagas avaliadas: {total_vagas}")
    print(f"Vagas com pelo menos 1 match: {vagas_com_match}")
    print(f"Total de candidatos corretos no Top-{k}: {total_matches}")

    if vagas_com_match == 0:
        print("\nNenhum candidato relevante foi recuperado no Top-K.")
        print("Sugestões:")
        print("- Verifique se os IDs de candidatos estão alinhados com os 'codigo' dos prospects.")
        print("- Reduza o threshold do ranqueamento supervisionado (ex: de 0.55 para 0.45).")
        print("- Verifique se os currículos estão suficientemente informativos.")
        print("- Teste manualmente uma vaga com candidatos reais e veja se aparecem no ranking.")
    else:
        print("\nRanking está recuperando candidatos relevantes.")
        print("Você pode prosseguir com o cálculo das métricas.")


# Executa o validador
validar_ranking(ranking_final, df_prospeccoes, df_applicants, k=5)
```
